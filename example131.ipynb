```python
%matplotlib nbagg
%config IPCompleter.greedy=True

from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import vc
import vplot


A = np.array([1,2,3]);
B = np.array([3,4,5]);

g = A - B;

l = vc.unit(A);
m = vc.unit(vc.cross(g,l));
n = vc.unit(vc.cross(l,m));

print(l);
print(m);
print(n);

print(vc.dot(m,l) );
print (vc.dot(m,n));
print(vc.dot(n,l) );

if (round(vc.dot(m,l)) == 0 and round(vc.dot(m,n),4) == 0 and round(vc.dot(n,l)) == 0):
	print("l,m,n perpendicular");
else:
	print("l,m,n  not perpendicular");



fig = plt.figure();
ax = plt.axes(projection='3d');	

vplot.plotvec(m, ax, 'r',3);
vplot.plotvec(l, ax, 'b',3);
vplot.plotvec(n, ax, 'g',3);

plt.show();
```
