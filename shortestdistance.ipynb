```python
%matplotlib nbagg

import vc
import vplot
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt

a = np.array([-2,2,1]);
c = np.array([5,1,2]);
b = np.array([10,0,3]);

p = a + vc.dot((c - a), vc.unit(b)) * vc.unit(b) - c; 

print("p magnitude: ", vc.magnitude(p))

print("cross check ", vc.dot(p,b));

fig = plt.figure()
ax = plt.axes(projection='3d')

vplot.plotvec(a, ax, 'r',2);
vplot.plotvec(c, ax, 'b',2);
vplot.plotVecUsePos(b,a,ax, 'g',2);
vplot.plotVecUsePos(p,c, ax, 'm',3);

plt.show()

```
