```python
%matplotlib nbagg
%config IPCompleter.greedy=True

from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import vc
import vplot


a = 5
h = 15

i = np.array([1, 0, 0]);
j = np.array([0, 1, 0]);
k = np.array([0, 0, 1]);

x = np.array([]);
y = np.array([]);
z = np.array([]);


for s in range(0, 100):
    r = a * np.cos(s/np.sqrt(a * a + h * h))*i + a * np.sin(s /np.sqrt(a*a + h * h))*j + h * s/np.sqrt(a * a + h * h) * k;
    x = np.append(x, r[0]);
    y = np.append(y, r[1]);
    z = np.append(z, r[2]);
    
drds = -a/np.sqrt(a * a + h * h) * np.sin(1/np.sqrt(a * a + h * h))*i + a / np.sqrt(a*a + h * h)* np.cos(1 /np.sqrt(a*a + h * h))*j + h / np.sqrt(a * a + h * h) * k;    
    
print('drds', drds);    
    
fig = plt.figure();
ax = plt.axes(projection='3d');

'''ax.scatter(x, y, z, c='r', marker='d')'''
ax.scatter(x, y, z, c='r', marker='^')
ax.plot(x,y,z, color='r')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

vplot.plotvec(drds * 5, ax, 'b', 3);


plt.show()


    

    
    
    


```
