```python
%matplotlib nbagg

import time
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import vc
import vplot



a = np.array([5,3,5]);
b = np.array([10,10,10]);

c = vc.dot(a,vc.unit(b)) * vc.unit(b);
d = vc.dot(a, b) * vc.unit(b);

print ("a ", a);
print ("b ", b);
print ("c ", c);
print ("d ", d);

fig = plt.figure();
ax = plt.axes(projection='3d');

#hint
#r, r--, ro--, ro, r+, bs--, bp--, g^
#plot(x, y, color='green', marker='p', linestyle='dashed',linewidth=2, markersize=12)

vplot.plotvec(a, ax, 'r',3);


#time.sleep(5);

vplot.plotvec(b, ax, 'b',3);


vplot.plotvec(c, ax, 'go--',1);
vplot.plotvec(d, ax, 'mo--',1);

plt.show();
```
