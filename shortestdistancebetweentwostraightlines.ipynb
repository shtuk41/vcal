```python
%matplotlib nbagg

import vc
import vplot
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt

a = np.array([-2,10,1]);
b = np.array([10,0,3]);
c = np.array([5,1,2]);
d = np.array([5,5,5]);

bu = vc.unit(b);
du = vc.unit(d);
pu = vc.cross(bu, du) / vc.magnitude(vc.cross(bu,du));

pmin = abs(vc.dot((a - c), pu));

print("p min: ", pmin);

fig = plt.figure()
ax = plt.axes(projection='3d')

vplot.plotVecUsePos(b,a,ax,'r',2);
vplot.plotVecUsePos(d,c, ax,'b',3);

plt.show()
```
