```python
%matplotlib nbagg

import vc
import vplot
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt

a = np.array([2,5,3]);
b = np.array([7,10,8]);
c = np.array([0,6,3]);
d = np.array([-12,0,9]);

bu = vc.unit(b - a);
du = vc.unit(d - c);
pu = vc.cross(bu, du) / vc.magnitude(vc.cross(bu,du));

pmin = abs(vc.dot((a - c), pu));

print("p min: ", pmin);

fig = plt.figure()
ax = plt.axes(projection='3d')

vplot.plotVecUsePos(b,a,ax,'r',0.8);
vplot.plotVecUsePos(d,c, ax,'b',1.0);

plt.show()
```
